First in first we neet to create our test data respect to latters

```python
#seperate data frame respect to latter,
latter_set = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
a_df = df[df["latter"] == "A"]
# learn data_length
num_row = len(a_df)
```

```python
chunk_size = 5
chunks = [a_df.iloc[i:i+chunk_size] for i in range(0, len(a_df), chunk_size)]

# YalnÄ±zca 5 satÄ±rlÄ± olan alt DataFrame'leri seÃ§elim
chunks = [chunk for chunk in chunks if len(chunk) == chunk_size]

chunks[0]
```

```python
import pandas as pd

# column_keys listesini tanÄ±mlayÄ±n
column_keys = [
    "gyroX_mean", "gyroY_mean", "gyroZ_mean",
    "accelX_mean", "accelY_mean", "accelZ_mean",
    "gyroX_std", "gyroY_std", "gyroZ_std",
    "accelX_std", "accelY_std", "accelZ_std",
    "target_column"
]

# SÃ¼tun adlarÄ±yla uyumlu olarak boÅŸ bir DataFrame oluÅŸturun
feature_df = pd.DataFrame(columns=column_keys)

# OluÅŸturulan DataFrame'i ekrana yazdÄ±rÄ±n
print(feature_df)

# chunks dÃ¶ngÃ¼sÃ¼ baÅŸlamadan Ã¶nce tanÄ±mlanmÄ±ÅŸ olmalÄ±dÄ±r
for i, chunk in enumerate(chunks, start=1):
    # Her bir alt DataFrame iÃ§in istatistiksel Ã¶zellikleri hesaplayÄ±n
    mean_values = chunk[['gyroX', 'gyroY', 'gyroZ', 'accelX', 'accelY', 'accelZ']].mean()
    std_values = chunk[['gyroX', 'gyroY', 'gyroZ', 'accelX', 'accelY', 'accelZ']].std()

    # Yeni satÄ±r verilerini oluÅŸturun
    new_row_data = {
        "gyroX_mean": mean_values["gyroX"],
        "gyroY_mean": mean_values["gyroY"],
        "gyroZ_mean": mean_values["gyroZ"],
        "accelX_mean": mean_values["accelX"],
        "accelY_mean": mean_values["accelY"],
        "accelZ_mean": mean_values["accelZ"],
        "gyroX_std": std_values["gyroX"],
        "gyroY_std": std_values["gyroY"],
        "gyroZ_std": std_values["gyroZ"],
        "accelX_std": std_values["accelX"],
        "accelY_std": std_values["accelY"],
        "accelZ_std": std_values["accelZ"],
        "target_column": df.iloc[1, 1]  # Burada df yerine chunk kullanmanÄ±z gerekebilir
    }

    # Yeni satÄ±rÄ± DataFrame'e ekleyin
    feature_df = pd.concat([feature_df, pd.DataFrame([new_row_data])], ignore_index=True)
```
